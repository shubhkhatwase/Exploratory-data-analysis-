```python
plt.subplot(221)
df['Maker'].value_counts().plot(kind='bar',title='Name of Car Maker companys',figsize=(40,26))
plt.xticks(rotation = 90)
```

```python
plt.subplot(222)
df['Model'].value_counts().plot(kind = 'bar',title = 'Models of Cars',figsize = (45,14))
plt.xticks(rotation = 90)
```

```python
plt.subplot(222)
df['Engine Type'].value_counts().plot(kind = 'bar',title = 'Soled Cars',figsize = (40,14))
plt.xticks(rotation = 90)
```

```python
plt.subplot(222)
df['Fuel Tank Capacity'].value_counts().plot(kind = 'bar',title = 'Fuel Tank Capacity of cars',figsize = (20,13))
plt.xticks(rotation = 90)
```

```python
plt.subplot(222)
df['Fuel Efficiency'].value_counts().plot(kind = 'bar',title = 'Soled Cars',figsize = (35,14))
plt.xticks(rotation = 90)
```

```python
df.boxplot()
```
