## Time Series & categorical Data
pandas has sample , powerful and effiecient functionallity for performing resampling operations during frequency conversion.

```python
date = pd.date_range(('20200301'),periods = 100,freq = 'S')
date
```

```python
dates = pd.date_range('3/3/2020',periods = 100,freq = 'S')
dates
```

```python
 ts = pd.Series(np.random.randint(0,500,len(dates)),dates)
 ts.resample('5min').sum()
```

```python
ts_utc = ts.tz_localize('UTC')
ts_utc 
```

```python
ts_utc.tz_convert('US/EASTERN')
ts_utc
```

```python
dates = pd.date_range('3/3/2020 00:00',periods = 100,freq = 'M')
ts = pd.Series(np.random.randn(len(dates)),dates)
ts
```

```python
ps = ts.to_period()
ps
```

```python
ps.to_timestamp()
```
